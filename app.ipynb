```
!pip install -Uqq fastbook
import fastbook
fastbook.setup_book()
from fastbook import *
from fastai.vision.widgets import *
path = Path("/content/gdrive/MyDrive/thesis/invasive_plants_II")

def parent_label_multi(o):
    return [Path(o).parent.name]
learn_inf = load_learner(path/'fmodel.pkl', cpu=True)
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()

lbl_pred = widgets.Label()

def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    labels, prediction, probability = learn_inf.predict(img)
    for idx,label in enumerate(learn_inf.dls.vocab):
      result = result + f"/n category:{label}, prediction:{prediction[idx]}, probablity:{probability[idx]}"
    lbl_pred.value = result
    
btn_run.on_click(on_click_classify)
VBox([widgets.Label('Select plant!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
